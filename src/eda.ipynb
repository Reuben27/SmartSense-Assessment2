## Analysis of Fields and Plots

Following are the questions asked as mentioned in codebook.txt
From what I understand, 
- E: Measure of Extrovertness
- N: Measure of Nervousness
- A: Measure of Friendliness
- C: Measure of Carelessness
- O: Open ended questions

As you can see, some questions are opposites to each other. So we need to change the score paramter (reverse it).

```python
E_questions = {
  'E1':  'I am the life of the party.', 
  'E2':  'I don\'t talk a lot.', 
  'E3':  'I feel comfortable around people.', 
  'E4':  'I keep in the background.', 
  'E5':  'I start conversations.',
  'E6':  'I have little to say.', 
  'E7':  'I talk to a lot of different people at parties.', 
  'E8':  'I don\'t like to draw attention to myself.',
  'E9':  'I don\'t mind being the center of attention.',
  'E10': 'I am quiet around strangers.'
}

N_questions = {
  'N1':  'I get stressed out easily.',
  'N2':  'I am relaxed most of the time.',
  'N3':  'I worry about things.',
  'N4':  'I seldom feel blue.', 
  'N5':  'I am easily disturbed.', 
  'N6':  'I get upset easily.', 
  'N7':  'I change my mood a lot.',
  'N8':  'I have frequent mood swings.',
  'N9':  'I get irritated easily', 
  'N10': 'I often feel blue.'
} 

A_questions = {
  'A1':  'I feel little concern for others.',
  'A2':  'I am interested in people.',
  'A3':  'I insult people.',
  'A4':  'I sympathize with others feelings.',
  'A5':  'I am not interested in other people problems.',
  'A6':  'I have a soft heart',
  'A7':  'I am not really interested in others.',
  'A8':  'I take time out for others', 
  'A9':  'I feel others emotions', 
  'A10': 'I make people feel at ease.'
}  

C_questions = {
  'C1': 'I am always prepared.', 
  'C2': 'I leave my belongings around.',
  'C3': 'I pay attention to details',
  'C4': 'I make a mess of things.', 
  'C5': 'I get chores done right away.', 
  'C6': 'I often forget to put things back in their proper place.', 
  'C7': 'I like order.', 
  'C8': 'I shirk my duties.', 
  'C9': 'I follow a schedule.',
  'C10': 'I am exacting in my work.'
}  

O_questions = {'O1':  'I have a rich vocabulary.', 
  'O2':  'I have difficulty understanding abstract ideas.',
  'O3':  'I have a vivid imagination.', 
  'O4':  'I am not interested in abstract ideas.',  
  'O5':  'I have excellent ideas', 
  'O6':  'I do not have a good imagination.',  
  'O7':  'I am quick to understand things.',
  'O8':  'I use difficult words.', 
  'O9':  'I spend time reflecting on things.',  
  'O10': 'I am full of ideas.'
} 
```

```python
reverse_values_dict = { 5:1 , 4:2 ,3:3 , 2:4, 1:5}
changed_qs = ['E2', 'E4', 'E6', 'E8', 'E10', 'N2', 'N4', 'A1', 'A3', 'A5', 'A7', 'C2', 'C4', 'C6', 'C8', 'O2', 'O4', 'O6']
for col in df.iloc[:, 7:]:
    if col in changed_qs:
      df.replace({col: reverse_values_dict})
    else:
      pass

df.head()
```

```python
plt.figure(figsize=[15, 25])
questions = E_questions.keys()
n = 1
for q in questions:
    plt.subplot(5, 2, n)
    sns.countplot(x=q, data=df)
    plt.title("Extrovertness \n " + E_questions[q])
    n=n+1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=[15, 25])
questions = N_questions.keys()
n = 1
for q in questions:
    plt.subplot(5, 2, n)
    sns.countplot(x=q, data=df, palette='BrBG')
    plt.title("Nervousness \n " + N_questions[q])
    n=n+1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=[15, 25])
questions = C_questions.keys()
n = 1
for q in questions:
    plt.subplot(5, 2, n)
    sns.countplot(x=q, data=df)
    plt.title("Carelessness \n " + C_questions[q])
    n=n+1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=[15, 25])
questions = A_questions.keys()
n = 1
for q in questions:
    plt.subplot(5, 2, n)
    sns.countplot(x=q, data=df, palette='BrBG')
    plt.title("Friendliness \n " + A_questions[q])
    n=n+1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=[15, 25])
questions = O_questions.keys()
n = 1
for q in questions:
    plt.subplot(5, 2, n)
    sns.countplot(x=q, data=df)
    plt.title("Openended \n " + O_questions[q])
    n=n+1
plt.tight_layout()
plt.show()
```

## Computing the total scores for each attribute

```python
df['O'] =  (df['O1'] + df['O2'] + df['O3'] + df['O4'] + df['O5'] + df['O6'] + df['O7'] + df['O8' ] + df['O9'] + df['O10']) / 10
df['C'] =  (df['C1'] + df['C2'] + df['C3'] + df['C4'] + df['C5'] + df['C6'] + df['C7'] + df['C8' ] + df['C9'] + df['C10']) / 10    
df['E'] =  (df['E1'] + df['E2'] + df['E3'] + df['E4'] + df['E5'] + df['E6'] + df['E7'] + df['E8' ] + df['E9'] + df['E10']) / 10
df['A'] =  (df['A1'] + df['A2'] + df['A3'] + df['A4'] + df['A5'] + df['A6'] + df['A7'] + df['A8' ] + df['A9'] + df['A10']) / 10
df['N'] =  (df['N1'] + df['N2'] + df['N3'] + df['N4'] + df['N5'] + df['N6'] + df['N7'] + df['N8' ] + df['N9'] + df['N10']) / 10
```

```python
new_df = df[['race', 'age', 'engnat', 'gender', 'hand', 'source', 'country', 'O','C', 'E','A','N']]
new_df.head()
```

```python
new_df.to_csv('data/BIG5/new.csv')
```
