## BIG 5

```python
df = pd.read_csv("data/BIG5/data.csv", sep='\t')
df.head()
```

```python
print("Shape:" , df.shape)
print("Columns:", list(df.columns))
print(len(list(df.columns)))
print(df.info())
```

```python
df.describe()
```

```python
reverse_values_dict = { 5:1 , 4:2 ,3:3 , 2:4, 1:5}
changed_qs = ['E2', 'E4', 'E6', 'E8', 'E10', 'N2', 'N4', 'A1', 'A3', 'A5', 'A7', 'C2', 'C4', 'C6', 'C8', 'O2', 'O4', 'O6']
for col in df.iloc[:, 7:]:
    if col in changed_qs:
      df.replace({col: reverse_values_dict})
    else:
      pass
  
```
