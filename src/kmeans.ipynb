```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

# Kmeans Clustering

Looking at the data, with 5 fields - O,C,E,A,N. The model that comes to my mind is Kmeans clustering. For now I will try to work with 5 clusters one for each.

```python
df = pd.read_csv("data/BIG5/data.csv", sep='\t')
reverse_values_dict = { 5:1 , 4:2 ,3:3 , 2:4, 1:5}
changed_qs = ['E2', 'E4', 'E6', 'E8', 'E10', 'N2', 'N4', 'A1', 'A3', 'A5', 'A7', 'C2', 'C4', 'C6', 'C8', 'O2', 'O4', 'O6']
for col in df.iloc[:, 7:]:
    if col in changed_qs:
      df.replace({col: reverse_values_dict})
    else:
      pass

df.head()
```

```python
df = df.drop(['race', 'age', 'engnat', 'gender', 'hand', 'source', 'country'], axis=1)
kmeans = KMeans(n_clusters=5, random_state=21)
k_fit = kmeans.fit(df)
predictions = k_fit.labels_
df['clusters'] = predictions
df.head()
```

```python
pca_2 = PCA(n_components=2)
pca_fit = pca_2.fit_transform(df)

df_pca = pd.DataFrame(data=pca_fit, columns=['PCA1', 'PCA2'])
df_pca['clusters'] = predictions
df_pca.head()
```
